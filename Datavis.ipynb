```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
all_month = pd.read_csv("all_month.csv")
all_month.head()
```

```python
all_month
```

```python
import matplotlib.pyplot as plt

# Create a histogram of earthquake magnitudes
plt.hist(all_month['mag'], bins=20, edgecolor='b')

plt.title('Histogram of Earthquake Data')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.legend()
```

```python
# Create a Bar chart- magnitudes

x = all_month.groupby(['place']).sum('mag')['mag'].to_frame('Magnitude').reset_index().sort_values('Magnitude',ascending=False).head(5)

plt.bar(x['place'], x["Magnitude"])
plt.title('Bar_chart of Earthquake Data')
plt.xlabel('place')
plt.ylabel('Frequency')
plt.xticks(rotation=75)
plt.legend()
```

```python
import matplotlib.pyplot as plt

# Group by 'depth', sum the 'mag' values, and select the top 5 depths by magnitude
top_5_depths = all_month.groupby('depth')['mag'].sum().nlargest(5)

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_depths, labels=top_5_depths.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Earthquake Depths by Magnitude')
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 5))

# Scatter Plot
ax.scatter(all_month['longitude'], all_month['latitude'], c='blue')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Earthquake Locations (Scatter Plot)')
```

```python
# Magnitude Frequency
mag = all_month['mag'].value_counts().sort_index()

# Line chart to visualising by magnitude
plt.figure(figsize=(8, 6))
plt.plot(mag.index, mag.values, marker='1', linestyle='-', color='green')
plt.title('Earthquakes by Magnitude')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.grid(True)  # Add grid lines for better readability
plt.show()
```

```python
## Depth Error by Location Source
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))

sns.boxplot(x='locationSource', y='depthError', data=all_month, palette='viridis')
plt.title('Depth Error by Location Source')
plt.xlabel('Location Source')
plt.ylabel('Depth Error')
plt.xticks(rotation=45)
location_mapping = {
    'us': 'Indonesia','ak' :'Morocco', 'nc' :'Kermadec Island', 'ci' :'Japan', 'hv' : 'Philippines'
}
# Replace the location column values based on the mapping
all_month['locationSource'] = all_month['locationSource'].replace(location_mapping)
plt.xticks(rotation=45)
plt.show()
```
